# Dataset loop

```python
import numpy as np
import nbformat
from nbconvert.preprocessors import ExecutePreprocessor
```

```python
notebook_path = "./MPC_evaluation_clean.ipynb"
```

Buildings to exclude:
2, 6, 7, 14, 15, 18, 19

```python
noise = [0.1,0.25,1,1.75,2.75,4]
houses = range(0,27)
```

## building

```python
# Read the notebook in
with open(notebook_path, "r") as f:
    notebook_content = f.read()
    for house in houses:
        # Replace the 'id' value for each iteration
        modified_notebook_content = notebook_content.replace("building = 0", f"building = {house}")

        # Create a temporary modified notebook
        modified_notebook = nbformat.reads(modified_notebook_content, as_version=nbformat.NO_CONVERT)

        # Execute the modified notebook
        execute_preprocessor = ExecutePreprocessor(timeout=-1)  # Set the timeout as needed
        try:
            execute_preprocessor.preprocess(modified_notebook)
        except:
            print(f"Error for building = {house}")
        print(f"Execution complete for building = {house}")
```

## Noise

```python
# Read the notebook in
with open(notebook_path, "r") as f:
    notebook_content = f.read()
    for n in noise:
        #value = round(n,2)
        value = n
        # Replace the 'id' value for each iteration
        modified_notebook_content = notebook_content.replace("noise = 0.5", f"noise = {value}")

        # Create a temporary modified notebook
        modified_notebook = nbformat.reads(modified_notebook_content, as_version=nbformat.NO_CONVERT)

        # Execute the modified notebook
        execute_preprocessor = ExecutePreprocessor(timeout=-1)  # Set the timeout as needed
        try:
            execute_preprocessor.preprocess(modified_notebook)
        except:
            print(f"Error for noise = {value}")
        print(f"Execution complete for noise = {value}")
```
