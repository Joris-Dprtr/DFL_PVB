```python
building = 0
price = 'Realistic'
noise = 0.5
```

```python
import src.data.dataprep as prep
from scipy.stats import skewnorm

def simulate_forecast(actual_load, noise_scale=0.1, skewness=-5):
    """
    Simulate forecast using a skewed noise distribution.

    Parameters:
    - actual_load: Series of actual values
    - noise_scale: scale (std dev) of noise applied as % of actual
    - skewness: negative values will skew the noise left (underestimation)

    Returns:
    - forecasted_load: Series of simulated forecasts
    """
    # Generate skewed noise from skew-normal distribution
    noise = skewnorm.rvs(a=skewness, size=len(actual_load)) * noise_scale

    # Apply noise and clip
    forecasted_load = actual_load * (1 + noise)
    forecasted_load = forecasted_load.clip(lower=0)

    return forecasted_load
```

```python
# Import the base data and resample it from 5 minutes to hourly
nl_data = prep.dutch_data('../data/Dutchdata_clean/building_' + str(building) + '.parquet', 'h', price=price)
```

```python
load_fcst = simulate_forecast(nl_data['load'], noise_scale=noise)
```

```python
load_fcst.to_csv('../data/load_fcsts/building_' + str(building) + '_' + str(noise) + '_noise_forecast.csv')
```
